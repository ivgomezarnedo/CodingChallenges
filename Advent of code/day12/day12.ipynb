```python
with open('input.txt') as f:
    lines = f.read().splitlines()
```

```python
lines
```

```python
direction=''
north_south=0 #North positive, south negative
east_west= 0 #East positive, west negative

for idx,line in enumerate(lines):
    next_direction=line[0]
    value=int(line[1:])
    if(next_direction=='W'):
        direction=next_direction
        east_west-=value
    elif(next_direction=='S'):
        direction=next_direction
        north_south-=value
    elif(next_direction=='E'):
        direction=next_direction
        east_west+=value
    elif(next_direction=='N'):
        direction=next_direction
        north_south+=value
    elif(next_direction=='F'):
        if(direction=='W'):
            east_west-=value
        elif(direction=='S'):
            north_south-=value
        elif(direction=='E'):
            east_west+=value
        elif(direction=='N'):
            north_south+=value
    elif(next_direction=='L'):
        if(direction=='W'):
            direction=next_direction
        elif(direction=='S'):
            direction=next_direction
        elif(direction=='E'):
            direction=next_direction
        elif(direction=='N'):
            direction=next_direction 
    elif(next_direction=='N'):
        direction='North'
        north_south+=value
        
    
```
